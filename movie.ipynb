```python
import pandas as pd
import tkinter as tk
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
# Load dataset
data2 = pd.read_csv("D:\\Office Data\\Downloads\\movie_recommender_dataset_300.csv")
data2
```

```python
# Features and target
X = data2[["Action_Rating", "Romance_Rating", "Comedy_Rating"]]
y = data2["Genre"]
```

```python
# Encode Genre
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42
)
```

```python

# Train model
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# Calculate accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
# ---------- Tkinter GUI ----------
def predict_genre():
    a = int(action_entry.get())
    r = int(romance_entry.get())
    c = int(comedy_entry.get())

    new_movie = [[a, r, c]]
    prediction = model.predict(new_movie)
    genre = le.inverse_transform(prediction)[0]

    result_label.config(text=f"Predicted Genre: {genre}")
```

```python
# Create window
root = tk.Tk()
root.title("Movie Genre Predictor")
root.geometry("380x350")
```

```python
# Labels & Entries
tk.Label(root, text="Action Rating (1-5)").pack(pady=5)
action_entry = tk.Entry(root)
action_entry.pack()

tk.Label(root, text="Romance Rating (1-5)").pack(pady=5)
romance_entry = tk.Entry(root)
romance_entry.pack()

tk.Label(root, text="Comedy Rating (1-5)").pack(pady=5)
comedy_entry = tk.Entry(root)
comedy_entry.pack()
```

```python
# Button
tk.Button(root, text="Predict Genre", command=predict_genre).pack(pady=15)

# Result
result_label = tk.Label(root, text="", font=("Arial", 12, "bold"))
result_label.pack()
```

```python
# Accuracy display
accuracy_label = tk.Label(
    root,
    text=f"Model Accuracy: {accuracy * 100:.2f}%",
    font=("Arial", 11)
)
accuracy_label.pack(pady=10)

root.mainloop()
```
